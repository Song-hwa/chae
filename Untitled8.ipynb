HSV 색상공간을 사용해서 원하는 색상만 출력하기

```python
import cv2
import numpy as np

img = cv2.imread('BALLOON.jpg')

#BGR색상을 HSV 색상으로 변환

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
#빨간색의 영역
#최소값
lower = np.array([-10,100,100])
#최대값
upper = np.array([10,255,255])

#최소값과 최대값의 범위에 해당하는 이미지를 선택
mask  = cv2.inRange(hsv, lower, upper)

#(이미지1, 이미지2, 적용 영역)
#이미지1, 이미지2를 비교 mask 값으로 비교 둘다 일치하면 출력
#아니면 0

img2 = cv2.bitwise_and(img, img, mask = mask)

cv2.imshow('red BALLOON', img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#침식
- 이진이미지 처리 방법
- Erosion(침식) : 이미지 경계부분에서 1을 0으로 변환하는 방법, 검은색을 확장
- dilation(팽창) : 이미지 경계부분에서 0을 1로 변환하는 방법, 흰색을 확장 
```

```python
import cv2
import numpy as np

# 흑백 이미지로 읽기
img  = cv2.imread('number.png',cv2.IMREAD_GRAYSCALE)

cv2.imshow('original', img)

#침식/ 팽창 크기를 설정
kernel = np.ones((3,3),np.uint8)

#침식
#이미지, 침식의 크기, 변환을 몇 번 할 것인지


erosion = cv2.erode(img, kernel, iterations = 1)

cv2.imshow('erosion', erosion)



#팽창
dilation = cv2.dilate(img, kernel, iterations = 1)

cv2.imshow('dilation', dilation)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('histogram.png',cv2.IMREAD_GRAYSCALE)

cv2.imshow('original', img)
#히스토그램 평활화
img2 = cv2.equalizeHist(img)

cv2.imshow('equalize', img2)

#이미지, 채널(흑백 0 , 컬러 : Blue(0), Green(1), Red(2))
#마스크 영역, 색상의 수, 픽셀 값의 범위
hist = cv2.calcHist([img2], [0], None, [256], [0, 255])

plt.plot(hist)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- 히스토그램 분포가 전체 픽셀에 대해 고르게 분포하도록 만드는 것(정규화)
- 흐릿한 이미지를 선명하게 만들 수 있음

```python
탬플릿 매칭
- 큰 이미지내에서 내가 원하는 작은 이미지를 찾기

```

```python
import cv2
import numpy as np
#배경 
# 결과 이미지용 컬러 사진
img1 = cv2.imread('game.png')

#사과

img2 = cv2.imread('apple.png', cv2.IMREAD_GRAYSCALE)

#배경을 GrayScale
#템플릿 매칭용

img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

#찾아야하는 이미지의 크기 계산
# h: 행, w : 열
h, w = img2.shape

#탬플릿 매칭
#배경 이미지, 찾을 이미지, 매칭 방법

res = cv2.matchTemplate(img1_gray, img2, cv2.TM_CCOEFF_NORMED)

loc = np.where(res>0.7)

#[::1-] : 첫번째 인덱스와 두번째 인덱스 값을 서로 바꾼
#height, width >  x, y
for pt in zip(*loc[::-1]):
    cv2.rectangle(img1, pt, (pt[0]+w, pt[1] + h), (0,0,255),2)
    
cv2.imshow('match', img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#zip() : 2 개  이상의 리스트나 튜플에서 동일위치에 있는 값들 끼리 묶어줌
#튜플로 변환

x = [1,2,3,4]

y = [5,6,7,8]

a = zip(x,y)


for pt in a :
    print(pt)
```

```python
#[::-1] : 위치를 서로 바꾸는 것
#*함수의 매개변수의 수가 정해져있지 않을 때 사용

loc = (np.array([1,2,3,4]),np.array([5,6,7,8]))

for pt in zip(*loc[::-1]):
    print(pt)
```

```python
#* : 함수의 매개변수의 수가 정해져있지 않을 때 사용
# def sum(a,b) : 
#return a +  b
def add(*num):
    return np.sum(num)

print(add(2,3))
print(add(2,3,4))
print(add(2,3,4,5,6))
```
